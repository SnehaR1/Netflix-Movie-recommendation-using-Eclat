```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Dataset_master.xlsx - Netflix Movie Recommendation Ec.csv',header=None)
data.head()
```

```python
data.shape
```

```python
transactions=[]
for i in range(0,7466):
  transactions.append([str(data.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_confidence=0.2,min_support=0.003,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
results
```

```python
def inspect(results):
  product1=[tuple(result[2][0][0])[0] for result in results]
  product2=[tuple(result[2][0][1])[0] for result in results]
  support=[result[1] for result in results]
  return list(zip(product1,product2,support))
result=pd.DataFrame(inspect(results),columns=["Product1","Product2","Support"])
```

```python
result
```

```python
result.nlargest(n=10,columns=["Support"])
```
